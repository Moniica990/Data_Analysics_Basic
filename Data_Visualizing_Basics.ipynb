```python
import plotly.express as px
```

```python
fig = px.box(df, x='class', y='fare', title='Fare by Class')
fig.update_layout(xaxis_title='Class', yaxis_title='Fare')
fig.show()
```

majority female travelled with high exxpense than men from the above diagram


```python
sns.scatterplot(data=df,x='age',y='fare',hue='sex')
plt.show()
```

```python
sns.pairplot(df[['age','fare','parch','sibsp']])
plt.show()
```

how much they are correlated is shown in heapmaps

```python
correlation_matrix=df[['age','fare','parch','sibsp']].corr()
correlation_matrix
```

```python
sns.heatmap(correlation_matrix,annot=True)
plt.show()
```
